```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# This will open the curve in a new window
%matplotlib

# Plotting on a single curve
x = range(100)
yRandDots = [np.random.random() for i in x]
yRandLines = [np.random.random() for i in x]
ySin = [np.sin(2*np.pi*i/100) for i in x]

plt.plot(x, yRandDots, 'o', label="random dots", color="b")
plt.plot(x, yRandLines, label="random lines", color="g")
plt.plot(x, ySin, label="sin", color="y")

plt.title("scatter- line and dots")
plt.xticks([i*10 for i in range(10)], ["a-" + str(i*10) for i in range(10)]) # X axis labels
plt.ylabel("my y axis")
plt.legend()
```

```python
# Plotting multiple figures sequentially (draw figure 1 first, then draw figure 2)
x = range(100)
y = range(100)
plt.figure()
plt.title(1)
plt.plot(x,y)

plt.figure()
plt.title(2)
plt.plot(x,y)
```

```python
# plotting multiple figures concurrently (you can define them in any order)
x = range(100)
y1 = range(100)
y2 = [i*i for i in range(100)]

fig1 = plt.figure()
fig2 = plt.figure()

fig1_ax1 = fig1.add_subplot(111)
fig1_ax1.plot(x, y1)

fig2_ax1 = fig2.add_subplot(111)
fig2_ax1.plot(x, y2)
```

```python
# multiple subplots in the same figure

x = range(100)
y = range(100)

fig = plt.figure()
fig.suptitle("cool super title")

# 22 represent that the subplots will be a grid of 2,2. 1 represnt the subplot position "Z" pattern
fig_ax1 = fig.add_subplot(221)
fig_ax1.plot(x,y)
fig_ax1.set_title(1)
fig_ax1.set_xlabel("xlabel")

fig_ax2 = fig.add_subplot(222)
fig_ax2.set_title("2")
fig_ax2.plot(x,y)

fig_ax3 = fig.add_subplot(223)
fig_ax3.plot(x,y)
fig_ax3.set_title("3")

fig_ax4 = fig.add_subplot(224)
fig_ax4.set_title("4")
fig_ax4.plot(x,y)

plt.tight_layout()
```
